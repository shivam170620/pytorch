```python
from ultralytics import YOLO
```

```python
import shutil, os
model = YOLO("yolov8n.pt") # yolo v8 nano pretrained model
```

```python
model.train(data = dataset.location + "/data.yaml", epochs = 5, imgsz = 640, batch = 16, workers = 2, device =0, 
            name = "yolov8n-plate-detection")

os.mkdir("saved_models", exist_ok = True)
shutil.copy("runs/detect/yolov8n-plate-detection/weights/best.pt", "saved_models/best.pt")
shutil.copy("runs/detect/yolov8n-plate-detection/weights/last.pt", "saved_models/last.pt")
print("Best and Last model weights have been saved to the 'saved_models' directory.")
```
